This notebook demonstrates the complete process of data cleaning, analysis, and visualization for NYC Transit Subway Entrance and Exit Data. We will generate meaningful insights to help plan an efficient CCTV deployment system.

### Step 1: Data Cleaning & Preparation
### Load the Dataset

```python

import pandas as pd

# Upload the Excel file (Use the file upload button on Colab)
from google.colab import files
uploaded = files.upload()

# Load the dataset
file_path = list(uploaded.keys())[0]  # Get the uploaded file name
dataset = pd.read_excel(file_path, sheet_name='Cleaned Data')

# Display the first few rows
dataset.head()
```

### Inspect the Dataset

```python

# Check basic information
dataset.info()

# Display column names
print(dataset.columns)
```

### Select Relevant Columns

```python

columns_to_keep = [
    "Division", "Line", "Station Name", "Station Latitude", "Station Longitude",
    "ADA", "Free Crossover", "North South Street", "East West Street",
    "Entrance Latitude", "Entrance Longitude"
]

cleaned_dataset = dataset[columns_to_keep]
cleaned_dataset.head()
```

### Handle Missing Values

```python

# Check for missing values
cleaned_dataset.isnull().sum()

# Remove rows with missing essential data
cleaned_dataset.dropna(subset=["Station Name", "Station Latitude", "Station Longitude"], inplace=True)
```

## Step 2: Data Analysis & Visualization

```python

# Count Entrances Per Station
entrance_count = cleaned_dataset.groupby('Station Name').size().reset_index(name='Entrance Count')
entrance_count.head()
```

```python

# Analyze ADA Accessibility Coverage
ada_coverage = cleaned_dataset.groupby(['Division', 'ADA']).size().reset_index(name='Count')
ada_coverage.head()
```

### Visualize Data

```python

import matplotlib.pyplot as plt

# Plot: Top 10 Stations by Number of Entrances
top_10 = entrance_count.sort_values(by='Entrance Count', ascending=False).head(10)
plt.figure(figsize=(12, 6))
plt.bar(top_10['Station Name'], top_10['Entrance Count'])
plt.title('Top 10 Stations by Number of Entrances')
plt.xticks(rotation=45)
plt.show()
```

```python

# Plot: ADA Accessibility Coverage by Division
ada_coverage_pivot = ada_coverage.pivot(index='Division', columns='ADA', values='Count').fillna(0)
ada_coverage_pivot.plot(kind='bar', figsize=(12, 6))
plt.title('ADA Accessibility Coverage by Division')
plt.xlabel('Division')
plt.ylabel('Count of Entrances')
plt.xticks(rotation=0)
plt.show()
```
